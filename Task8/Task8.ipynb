```python
from py2neo import Graph

g = Graph('bolt://localhost:7687')
```

Tìm những bộ phim cùng thể loại với bộ phim “Kung Fu Panda 3” sắp xếp giảm theo số thể loại phim của các bộ phim đó.

```python
g.run("MATCH (m:Movie {title: 'Kung Fu Panda 3'})-[:IN_GENRE]->(g:Genre)<-[:IN_GENRE]-(m2:Movie) \
RETURN m2.title, COLLECT(g.name) AS DS \
ORDER BY size(DS) DESC LIMIT 20").data()
```

2. Cho biết những bộ phim cùng thể loại với thể loại các bộ phim mà 'Kim Brooks' đã đánh giá.

Chỉ lấy những bộ phim mà người này chưa đánh giá và sắp xếp các bộ phim theo thứ tự giảm của số thể loại.

```python
g.run("MATCH (u:User {name: 'Kim Brooks'})-[:RATED]->(m:Movie)-[:IN_GENRE]->(g:Genre)<-[:IN_GENRE]-(m2:Movie) \
MATCH (m2)-[:IN_GENRE]->(g2:Genre) WHERE NOT EXISTS {(m)-[:IN_GENRE]->(g2)} \
RETURN m2.title, COLLECT(DISTINCT g2.name) AS DS \
ORDER BY SIZE(DS) DESC").data()
```

Gợi ý những bộ phim cho ‘Kim Brooks’ theo đạo diễn phim.

```python
g.run('MATCH (u:User {name: "Kim Brooks"})-[:RATED]->(m:Movie)<-[:DIRECTED]-(d:Director)-[:DIRECTED]->(m2:Movie) \
RETURN m2.title, d.name').data()
```

4. Gợi ý những bộ phim theo mức độ rating của người xem: gợi ý cho ‘Kim Brooks’ những bộ phim có trung bình rating cao hơn hoặc bằng trung bình rating của ‘Kim Brooks’.

```python
g.run("MATCH (u:User {name: 'Kim Brooks'})-[r:RATED]->(m:Movie) \
WITH m, AVG(r.rating) AS AvgRatingKim \
MATCH (u2:User)-[r2:RATED]->(m2:Movie) \
WITH m, m2, AvgRatingKim, AVG(r2.rating) AS AvgRating \
WHERE AvgRating >= AvgRatingKim \
RETURN m2.title, AvgRating").data()
```

5. Gợi ý những bộ phim cho ‘Kim Brooks’ với cách xác định người tương tự:

a. Có chung rating ít nhất 10 bộ phim

```python
g.run("MATCH (u:User {name: 'Kim Brooks'})-[:RATED]->(m:Movie)<-[:RATED]-(u2:User) \
WITH u, u2, COUNT(m) AS SL \
WHERE SL >= 10 \
MATCH (u2)-[:RATED]->(m2:Movie) \
WHERE NOT (u)-[:RATED]->(m2) \
RETURN DISTINCT m2.title").data()
```

b. Lấy 10 người có số bộ phim rating chung lệch nhau <2 nhiều nhất

```python
g.run("MATCH (u:User {name: 'Kim Brooks'})-[r1:RATED]->(m:Movie)<-[r2:RATED]-(peer:User) \
WHERE abs(r1.rating-r2.rating) < 2  \
WITH u, peer, COUNT(m) AS SL ORDER BY SL DESC LIMIT 10 \
MATCH (peer)-[:RATED]->(m2:Movie) \
WHERE NOT (u)-[:RATED]->(m2) \
RETURN DISTINCT m2.title").data()
```

6. Gợi ý những bộ phim cho ‘Kim Brooks’ dùng K-NN với K = 20, sử dụng độ đo cosin

```python
g.run('MATCH (p1:User {name: "Kim Brooks"})-[x:RATED]->(m:Movie)<-[y:RATED]-(p2:User) \
WITH p1, p2, count(m) AS numbermovies, sum(x.rating * y.rating) AS xyDotProduct, \
collect(x.rating) as xRatings, collect(y.rating) as yRatings \
WHERE numbermovies > 10 \
WITH p1, p2, xyDotProduct, \
sqrt(reduce(xDot = 0.0, a IN xRatings | xDot + a^2)) AS xLength, \
sqrt(reduce(yDot = 0.0, b IN yRatings | yDot + b^2)) AS yLength \
WITH p1, p2, xyDotProduct / (xLength * yLength) AS sim \
ORDER BY sim DESC \
LIMIT 20 \
MATCH (p2)-[:RATED]->(m2:Movie) \
WHERE NOT (p1)-[:RATED]->(m2) \
RETURN DISTINCT m2.title').data()
```

7. Gợi ý những bộ phim cho người xem theo độ tương tự hai bộ phim:


a. Hai bộ phim tương tự nhau nếu có hơn 10 người người rating chung với độ lệch rating < 2, ưu tiên bộ phim có số người rating nhiều hơn.

```python
g.run('MATCH (m1:Movie {title: "Crimson Tide"})<-[r1:RATED]-(u:User)-[r2:RATED]->(m2:Movie) \
WITH m1, m2, COUNT(DISTINCT u) AS CountCommonUser, abs(r1.rating - r2.rating) AS DL \
WHERE CountCommonUser >= 10 AND DL < 2 \
RETURN m1.title, m2.title, CountCommonUser, DL \
ORDER BY CountCommonUser DESC LIMIT 20').data()
```

b. Hai bộ phim là tương tự nhau nếu có ít nhất 10 người rating chung và độ đo cosin của hai bộ phim. Lấy K bộ phim có độ tương tự cao nhất.

```python
g.run('MATCH (m1:Movie {title: "Crimson Tide"})<-[x:RATED]-(u:User)-[y:RATED]->(m2:Movie) \
WITH m1, m2, count(u) AS CommonUser, sum(x.rating * y.rating) AS xyDotProduct, \
collect(x.rating) as xRatings, collect(y.rating) as yRatings \
WHERE CommonUser > 10 \
WITH m1, m2, xyDotProduct, \
sqrt(reduce(xDot = 0.0, a IN xRatings | xDot + a^2)) AS xLength, \
sqrt(reduce(yDot = 0.0, b IN yRatings | yDot + b^2)) AS yLength \
RETURN m1.title, m2.title, xyDotProduct / (xLength * yLength) AS sim \
ORDER BY sim DESC \
LIMIT 20').data()
```
